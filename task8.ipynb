# Ordinal encoder

```python
df['income'].unique()
```

```python
income_order = ['low','medium','high']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enco = OrdinalEncoder(categories=[income_order])
```

```python
enco.fit(df[['income']])
```

```python
final = pd.DataFrame(enco.transform(df[['income']])) # n_samples , n_values
```

```python
df['income'] = final
```

```python
df
```

```python
df['age'] = final
```

```python
df['gender','m_status','buys'] = final
```
