```python
df = pd.read_csv("home data.csv")
df.head()
```

```python
df.shape
```

```python
x = df['x']
y = df['y']
```

# Gradient descent in ml

```python
#first step
m = 0 
c = 0
Learning_rate = 0.001

n = int(len(x))

iterations = 1000

#2nd step
for i in range(iterations):
    y_prediction = m * x + c
    dervatives_m = (-2/n) * sum(x*(y-y_prediction))
    dervatives_c = (-2/n) * sum((y-y_prediction))
    
    #final step
    m = m - Learning_rate * dervatives_m
    c = c - Learning_rate * dervatives_c
    print(m,c)
print('This is m:' ,m)
print('This is c:' ,c)
```
